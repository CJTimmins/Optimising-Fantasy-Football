```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
from scipy.stats import ttest_ind
from scipy.stats import shapiro
from scipy.stats import kstest
import array
import random
import numpy
import matplotlib
import matplotlib.pyplot as plt

from deap import algorithms
from deap import base
from deap import creator
from deap import tools
```

```python
# THIS FUNCTION READS THE DATA FILE CONTAINING THE INFORMATION RE EACH PLAYER

# read data

data = (pd.read_csv("clean-data.csv")
        .reset_index(drop=True))

num_players = len(data.index)

print("num possible players is %s" % (num_players))
```

```python
# HELPFUL DATA 
# these can be used for calculating points and costs and are also used in the constraint_checking function
points = data['Points'] 
cost = data['Cost']
    
# create lists with all elements initialised to 0
gk = np.zeros(num_players)
mid = np.zeros(num_players)
defe = np.zeros(num_players)
stri = np.zeros(num_players)

for i in range(num_players):
    if data['Position'][i] == 'GK':
        gk[i] = 1
    elif data['Position'][i] == 'DEF':
        defe[i] = 1
    elif data['Position'][i] == 'MID':
        mid[i] = 1
    elif data['Position'][i] == 'STR':
        stri[i]=1
  
```

# Algorithm

```python
#constraints

MAX_VAL = 100
TEAMSIZE = 11
TOTAL_VAL =0
for i in range(len(cost)):
    TOTAL_VAL+=cost[i]



    
POPSIZE=250
MUTPB=0.2
CXPB=0.8
NGEN=200
TOURNSIZE=3
```

```python
def initTeam(icls,size,numPlayers):
    
    #Value is not considered in intitialisation
    size=11
    #list of constraints
    #Val must be <=100
    #Size =11
    #All unique
    #1 GK
    #3>=DEF<=5
    #3>=MID=<=5
    #1>=STR<=3
    
    while(True):
        ind = icls(numpy.zeros(size))
        current_val=0
        item_indices=list(range(0,numPlayers))
        random.shuffle(item_indices)
        i=0
        GK=0
        DEF=0
        MID=0
        STR=0

        for pos in range(numPlayers):
            if (i==12):
                break
            player = item_indices[pos]
            playerVal = cost[pos]
            if (playerVal + current_val) <= MAX_VAL: #Budget check
                if (player not in ind): #unique check
                    #search all lists to find player role
                    goalkeeper = gk[player]
                    defender = defe[player]
                    midfielder = mid[player]
                    striker = stri[player]
                    if (goalkeeper==1):
                        if (GK==0):
                            ind[i]=player
                            current_val +=playerVal
                            GK+=1
                            i+=1
                    elif (defender==1):
                        if (DEF<=5):
                            ind[i]=player
                            current_val +=playerVal
                            DEF+=1
                            i+=1
                    elif(midfielder==1):
                        if (MID <=5):
                            ind[i]=player
                            current_val +=playerVal
                            MID+=1
                            i+=1
                    elif(striker==1): 
                        if (STR<=3):
                            ind[i]=player
                            current_val +=playerVal
                            STR+=1
                            i+=1
        if current_val<=MAX_VAL:
            break
            
        
    return ind

    
```

```python
def uniqList(individual):
    s = set()
    for x in individual:
        if x in s: return False
        s.add(x)
    return True
```

```python
def countNonUniq(individual):
    s = set()
    for x in individual:
        if x not in s:
            s.add(x)
    return len(individual)-len(s)
```

```python
def evalKnapsack(individual):
    value = 0.0
    pointTotal =0.0
    GK=0.0
    DEF=0
    MID=0
    STR=0
    fail=0
    overcost=False
    dist=1
    diff=1
    failDist=0
    for item in range(TEAMSIZE):
        value += cost[individual[item]]
        pointTotal += points[individual[item]]
        GK+=gk[int(individual[item])]
        DEF+=defe[int(individual[item])]
        MID+=mid[int(individual[item])]
        STR+=stri[int(individual[item])]
    if  value > MAX_VAL:
        overcost=True          # Value of over 100 get zero
    if not(uniqList(individual)):
        fail += countNonUniq(individual)
    if GK>1:
        fail+=(GK-1)
    if DEF<3 or DEF>5:
        if DEF<3:
            fail+=abs(DEF-3)
        elif DEF>5:
            fail+=(DEF-5)
    if MID<3 or MID>5:
        if MID<3:
            fail+=abs(DEF-3)
        elif MID>5:
            fail+=(MID-5)
    if STR<1 or STR>3:
        if STR<1:
            fail+=abs(STR-1)
        elif STR>5:
            fail+=STR-3           
    
    if fail>0 or overcost==True:
        if overcost==True:
            #calc distance in cost
            dist= abs(value-MAX_VAL)
            diff = min(MAX_VAL,abs(TOTAL_VAL-MAX_VAL))
        if fail>0:
            failDist=fail        
        P = pointTotal * (dist+failDist/diff+(failDist*(10)))
        #print("pointTotal: %s, failDist: %s, dist: %s, diff: %s, P: %s" % (pointTotal,failDist,dist,diff,P))
        pointTotal= pointTotal-P

    return  pointTotal,
```

```python
def mutRandomPlayer(individual):
    #Flip one element while keeping list unique
    
    
    potential = list(range(0,num_players))
    potential = [x for x in potential if x not in individual]
    remove = random.randint(0,len(individual)-1)
    if (uniqList(individual)):
        while(True):
            p = random.randint(0,len(potential)-1)
            candidate= potential[p]
            individual[remove]=candidate        
            if (uniqList(individual)):
                break
    else:       
        p = random.randint(0,len(potential)-1)
        candidate= potential[p]
        individual[remove]=candidate
    
    return individual,
```

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

# create a toolbox
toolbox = base.Toolbox()

# Attribute generator
toolbox.register("attr_bool", random.randint, 0, num_players)

toolbox.register("individual", initTeam, creator.Individual, 11,num_players)

#  a population consist of a list of individuals
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
def main():
    
    # choose a population size: e.g. 200
    pop = toolbox.population(n=POPSIZE)
    
    # keep track of the single best solution found
    hof = tools.HallOfFame(1)
 
    # create a statistics object: we can log what ever statistics we want using this. We use the numpy Python library
    # to calculate the stats and label them with convenient labels
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    stats.register("avg", numpy.mean)
    stats.register("std", numpy.std)
    stats.register("min", numpy.min)
    stats.register("max", numpy.max)
    
    # run the algorithm: we need to tell it what parameters to use
    # cxpb = crossover probability; mutpb = mutation probability; ngen = number of iterations
    pop, log = algorithms.eaSimple(pop, toolbox, cxpb=CXPB,mutpb=MUTPB, ngen=NGEN, 
                                   stats=stats, halloffame=hof, verbose=False)
    
    return pop, log, hof


```

```python
# register all operators we need with the toolbox
toolbox.register("evaluate", evalKnapsack)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", mutRandomPlayer)
toolbox.register("select", tools.selTournament, tournsize=TOURNSIZE)
```

```python
##############################
# run the main function 
pop, log, hof = main()

##############################


best = hof[0].fitness.values[0]   # best fitness found is stored at index 0 in the hof list


# look in the logbook to see what generation this was found at

max = log.select("max")  # max fitness per generation stored in log

for i in range(200):  # set to ngen
        fit = max[i]
        if fit == best:
            break        
        
print("max fitness found is %s at generation %s" % (best, i))
print(hof[0])
```

# Solution and constraint checker function

You are free to represent an individiual in any way you wish. However, at the end of the evolutionary run, you *must* convert your solution to a list of length *num_players* in which each element is either 0 or 1. An element *i* should be set to 0 if player *i* is not included in the team, and to 1 if player *is* **is** included in the team.

```python
# check the constraints
# the function MUST be passed a list of length num_players in which each bit is set to 0 or 1


def check_constraints(individual):
     
    broken_constraints = 0

    # exactly 11 players
    c1 = np.sum(individual)
    if  c1 != 11:
        broken_constraints+=1
        print("total players is %s " %(c1))
        
    
    #need cost <= 100"
    c2 = np.sum(np.multiply(cost, individual)) 
    if c2 > 100:
        broken_constraints+=1
        print("cost is %s " %(c2))
    
    # need only 1 GK
    c3 = np.sum(np.multiply(gk, individual))
    if  c3 != 1:
        broken_constraints+=1
        print("goalies is %s " %(c3))
    
    # need less than 3-5 DEF"
    c4 = np.sum(np.multiply(defe,individual))
    if  c4 > 5 or c4 < 3:
        broken_constraints+=1
        print("DEFE is %s " %(c4))
            
    #need 3- 5 MID
    c5 = np.sum(np.multiply(mid,individual))
    if  c5 > 5 or c5 < 3: 
        broken_constraints+=1
        print("MID is %s " %(c5))
        
    # need 1 -1 3 STR"
    c6 = np.sum(np.multiply(stri,individual))
    if c6 > 3 or c6 < 1: 
        broken_constraints+=1
        print("STR is %s " %(c6))
        
    # get indices of players selected
    selectedPlayers = [idx for idx, element in enumerate(individual) if element==1]
    
    totalpoints = np.sum(np.multiply(points, individual))
        
        
    print("total broken constraints: %s" %(broken_constraints))
    print("total points: %s" %(totalpoints))
    print("total cost is %s" %(c2))
    print("selected players are %s" %(selectedPlayers))
    
    return broken_constraints, totalpoints
```
